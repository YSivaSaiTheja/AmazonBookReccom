Setup+Prep

```python
import pandas as pd
import pickle
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
books_df = pd.read_csv("Books.csv")
ratings_df = pd.read_csv("Ratings.csv")
users_df = pd.read_csv("Users.csv")
```

Get Books_v2 dataset

```python
with open("books_data.pickle", "rb") as file:
    books_data = pickle.load(file)

# Clean the dictionary

for isbn, book_data in books_data.items():
    # Clean the genres
    genres_str = book_data["Genres"]

    if genres_str is None:
        genres_lst = []
        continue

    if "...more" in genres_str:
        genres_str = genres_str.split("...more")[0]

    genres_lst = genres_str.split("\n")[:-1]

    if len(genres_lst) == 1:
        genres_lst = []
    else:
        genres_lst = genres_lst[1:]

    books_data[isbn]["Genres"] = genres_lst

    # Clean the descriptions (remove the \n)
    desc_str = book_data["Description"]
    desc_str = desc_str.replace("\n", " ")
    books_data[isbn]["Description"] = desc_str

    # Clean the series
    series_str = book_data["Series"]
    # Check if part of a series
    if "#" not in series_str:
        series_str = None
    books_data[isbn]["Series"] = series_str
```

```python
books_data_df = pd.DataFrame(books_data).T
books_df = books_df.merge(books_data_df, left_on='ISBN', right_index=True)
books_df
```

```python
# books_df.to_csv("Books_v2.csv", index = False)
```

Clean the Datasets

```python
# Only keep books with descriptions and genres
books_df2 = books_df.dropna(subset=['Description', 'Genres'])
# Drop useless columns like image url
books_df2 = books_df2.drop(columns = ["Image-URL-S", "Image-URL-M", "Image-URL-L"])
books_df2
```

```python
# Create new ratings dataframe by filtering ratings dataframe
# to only include books which are in our smaller books dataset

isbn_lst = books_df2["ISBN"]
ratings_filtered_df = ratings_df[ratings_df['ISBN'].isin(isbn_lst)]
ratings_filtered_df = ratings_filtered_df.reset_index()
ratings_filtered_df = ratings_filtered_df.drop("index", axis = 1)
ratings_filtered_df
```

```python
# Get ratings count for each book
book_ratings_count = ratings_filtered_df.groupby("ISBN").count()
book_ratings_count.drop("Book-Rating", axis = 1, inplace = True)
book_ratings_count.rename({"User-ID": "Total_Nr_of_Ratings"}, axis = 1, inplace = True)
book_ratings_count

# Merge in book ratings count with the smaller books dataset
books_df3 = books_df2.merge(book_ratings_count, left_on="ISBN", right_index = True)
books_df3
```

```python
# Get country column for users dataframe
users_df["Country"] = users_df["Location"].str.split(", ").str[-1]
users_df
```

Get the Final Cleaned CSVs

```python
ratings_filtered_df.to_csv("Ratings_v2.csv", index = False)
```

```python
users_df.to_csv("Users_v2.csv", index = False)
```

```python
mlb = MultiLabelBinarizer(sparse_output=True)
books_df4 = books_df3.copy()
books_df4 = books_df4.join(
            pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(books_df4.pop('Genres')),
                index=books_df4.index,
                columns=mlb.classes_))
```

```python
books_df4.to_csv("Books_v3.csv", index = False)
```
